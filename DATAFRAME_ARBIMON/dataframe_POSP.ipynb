```python
import pandas as pd
```

```python
POSP_tp = pd.read_csv('rois.csv')
POSP_fp = pd.read_csv('rois.csv')
```

```python
#FILTRAR SOLO LAS ETIQUETAS VALIDADAS COMO 'present' y 'not present'
POSP_tp = POSP_tp[POSP_tp.validated == 'present']
POSP_fp = POSP_fp[POSP_fp.validated == 'not present']
```

```python
POSP_fp
```

```python
#REINICIAR INDICE
POSP_tp = POSP_tp.reset_index()
POSP_fp = POSP_fp.reset_index()
POSP_fp
```

```python
#Borrar columna que se crea automaticamente con reset_index
POSP_tp.drop(['index'],axis=1,inplace=True)
POSP_fp.drop(['index'],axis=1,inplace=True)
POSP_fp
```

```python
#Exportar csv
POSP_tp.to_csv('POSP_tp.csv')
POSP_fp.to_csv('POSP_fp.csv')
```
