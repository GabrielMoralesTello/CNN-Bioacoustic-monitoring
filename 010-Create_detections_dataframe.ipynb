```python
import os
import pickle
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.dates as mdates
from matplotlib import cm
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
import numpy as np
import pandas as pd
import time
import datetime
import seaborn as sns
from nltk import flatten
%matplotlib notebook
```

```python
recordings_wav = '/media/gabsoni/GABOWET/'
report_path = '../report/'
```

```python
path = '../binarios/'
sites = [s for s in sorted(os.listdir(path)) if not s.startswith('.')]
sites
```

```python
classes = [c for c in sorted(os.listdir(path+sites[0]+'/')) if not c.startswith('.') and c.endswith('presences.pkl')]
classes
```

```python
prob_class = [c for c in sorted(os.listdir(path+sites[0]+'/')) if not c.startswith('.') and c.endswith('probabilities.pkl')]
prob_class
```

```python
id_number = []
recording = []
wetland = []
species = []
year = []
month = []
day = []
hour = []
minute = []
x1 = []
x1_sec = []
probability = []
path_file = []

for i, clase in enumerate(classes):
    for j, site in enumerate(sites):
        
        data = open(path+site+'/'+clase, 'rb') 
        date_time = pickle.load(data)
        data.close()
        
        data2 = open(path+site+'/'+prob_class[i], 'rb') 
        prob = pickle.load(data2)
        data2.close()
        
        date_time = flatten(date_time)
        prob = flatten(prob)
        
        for k, stamp in enumerate(date_time):
            
            id_number.append(str(i)+str(j)+str(k))
            wetland.append(site)
            species.append(clase.split('_')[0]+'_'+clase.split('_')[1])
            year.append(stamp.year)
            month.append(stamp.month)
            day.append(stamp.day)
            hour.append(stamp.hour)
            minute.append(stamp.minute)
            x1.append(stamp.second)
            x1_sec.append(stamp.minute*60 + stamp.second)
            probability.append(prob[k])
            
            if len(str(stamp.month)) == 1:
                mm = '0' + str(stamp.month)
            else:
                mm = str(stamp.month)
            
            if len(str(stamp.day)) == 1:
                dd = '0' + str(stamp.day)
            else:
                dd = str(stamp.day)
            
            if len(str(stamp.hour)) == 1:
                hh = '0' + str(stamp.hour)
            else:
                hh = str(stamp.hour)
            filename = 'MON_'+ str(stamp.year) + mm + dd + '_' + hh + '0000.wav'    
            recording.append(filename)             
            path_file.append(recordings_wav + site + '/' + filename)

data = {'id_number': id_number, 'recording': recording, 'wetland': wetland, 'species': species,
        'year': year, 'month': month, 'day': day, 'hour': hour, 'minute': minute, 'x1': x1, 'x1_sec': x1_sec, 
        'probability': probability, 'path_file': path_file}

labels = pd.DataFrame(data=data)
labels.to_csv(report_path + 'detections_dataframe.csv')
labels
```
